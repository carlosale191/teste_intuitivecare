```python
import pdfplumber
import pandas as pd

dados = []
other_value = "N/A"
```

```python
#abre o arquivo, extrai as tabelas do arquivo, itera sobre cada tabela depois sobre cada linha
with pdfplumber.open("meus_pdfs/Anexo_I.pdf") as pdf:
    for page in pdf.pages:
        tabelas = page.extract_tables()
        for tabela in tabelas:
            for linha in tabela:
                #muda o valor NaN para outro em todas linhas com list comprehension
                alt_linha = [other_value if v == "NaN" else v for v in linha]
                dados.append(linha)
```

```python
#cria um dataframe
df = pd.DataFrame(dados)

#salva em csv
df.to_csv("tabelas_rol.csv", index=False, header=False, encoding="utf-8")
```
